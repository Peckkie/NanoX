# ▶️ Fig.2 AUC curve /threshold 0-1

```python
path = '/home/yupaporn/codes/NanoX/nanoX.csv'
import pandas as pd
df = pd.read_csv(path)
print(df.shape)
df.head()
```

```python
prop = df['Delta_C'].tolist()
Class = df['Label_'].tolist()
```

```python
import numpy as np
threshold = np.arange(0, 1.0001, 0.0001).tolist()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
resdf = []
TP, FP, TN, FN, Acc, Recall, Precision, TPR, FPR, SPEC, FNR, WL= [], [], [], [], [], [], [], [], [], [], [], []
for j in threshold:
    pre_ = []
    for i in prop:
        if i  >= j:
            pre_.append(1)
        else:
            pre_.append(0) 
    tn, fp, fn, tp = confusion_matrix(Class, pre_).ravel()
    print(j,'\n', tn, fp, fn, tp )
    pre = tp/(tp+fp)
    re= tp/(tp+fn)
    acc = (tp+tn)/(tp+tn+fp+fn)
    tpr = tp/(tp+fn)
    fpr = fp/(fp+tn)
    sen = tp/(tp+fn)
    spec = tn/(tn+fp)
    fnr = 1-(tp/(tp+fn)) 
    wl = (fp+tp)/(tp+tn+fp+fn)
    resdf.append({'Threshold' : j,'TP' : tp,'FP' : fp,'TN' : tn,'FN' : fn, 'Acc' : acc, 'TPR' : tpr, 'FPR' : fpr, 'FNR' : fnr, 'SPEC' : spec, 'SEN' : sen,'WL' : wl, 'Pre' : pre, 'Recall' : re})
```

# ▶️ Fig.3 Leave-one-out cross validation


```python
path = '/home/yupaporn/codes/NanoX/nanoX.csv'
import pandas as pd
df = pd.read_csv(path)
print(df.shape)
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import LeaveOneOut,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
##SUMMARY
X = df[['Delta A']]
Y = df['Label_']
    #train model ---
loo=LeaveOneOut()
model =LogisticRegression(solver='liblinear', C=10.0, random_state=0).fit(X, Y)
intercept = model.intercept_[0]
coef = model.coef_[0][0]

print(intercept)
print(coef)
```

```python
resdf = []
for i in range(1,44):
    a= 'set'+str(i)
    test = df[df[a] == 'test']
    train = df[df[a] != 'test']
    
    X = train[['Delta A']]
    Y = train['Label_']
    
    #train model ---
    loo=LeaveOneOut()
    model =LogisticRegression(solver='liblinear', C=10.0, random_state=0).fit(X, Y)
    intercept = model.intercept_[0]
    coef = model.coef_[0][0]
    
    #predict model ---
    x= test[['Delta A']]
    y= test['Label_']
    proba = model.predict_proba(x)
    pred = model.predict(x)[0]
    
    
    if pred == list(y)[0]:
        
        if  list(y)[0] == 0:
            con = 'TN'
        else:
            con = 'TP'
    else:
        if  list(y)[0] == 0:
            con = 'FP'
        else:
            con = 'FN'
    
    res = model.score(x, y)
    x = x.to_numpy()[0][0]
    resdf.append({ 'set' : a,'test(Delta A)' : x,'Actual label' : list(y)[0], 'intercept' : intercept, 'coef' : coef, 'prob' : proba, 'pred' : pred, 'result' : res, 'result_' : con})
```

```python
res_df = pd.DataFrame(resdf)
res_df
```

```python
TN = len(res_df[res_df['result_']== 'TN'])
TP = len(res_df[res_df['result_']== 'TP'])
FP = len(res_df[res_df['result_']== 'FP'])
FN = len(res_df[res_df['result_']== 'FN'])
```

```python
PPV = TP/(TP+FP)
NPV = TN/(TN+FN)
ACC = (TP+TN)/(TP+TN+FP+FN)
SPEC = TN/(TN+FP)
SEN = TP/(TP+FN)
print(f'PPV: {PPV} \nNPV: {NPV} \nACC: {ACC} \nSPEC: {SPEC} \nSEN: {SEN} ')
```
